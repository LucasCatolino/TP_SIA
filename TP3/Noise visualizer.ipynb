```python
import matplotlib.pyplot as plt
```

```python
originalError= []
out= []

i= 0

f = open("errorOriginal", "r")
for x in f:
    line = x.split(',')
    expected= float(line[0])
    real= line[1].rstrip("\n")
    error= expected - float(real)
    originalError.append(abs(error))
    out.append(i)
    i= i+1

f.close()
```

```python
noiseError= []

f = open("errorNoise", "r")
for x in f:
    line = x.split(',')
    expected= float(line[0])
    real= line[1].rstrip("\n")
    error= expected - float(real)
    noiseError.append(abs(error))

f.close()
```

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(20)

plt.plot(out, originalError, label = "Original")
plt.plot(out, noiseError, label = "Noise")
plt.legend()

plt.show()
```
