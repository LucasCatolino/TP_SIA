```python
import matplotlib.pyplot as plt
```

```python
epochs= []
trainAcc= []
f = open("accuracyTraining", "r")
for x in f:
    line = x.split(',')
    epochs.append(int(line[0]))
    acc= line[1].rstrip("\n")
    trainAcc.append(float(acc))

f.close()
```

```python
testAc= []
f = open("accuracyTesting", "r")
for x in f:
    line = x.split(',')
    acc= line[1].rstrip("\n")
    testAc.append(float(acc))

f.close()
```

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(20)

plt.plot(epochs, trainAcc, label = "Training")
plt.plot(epochs, testAc, label = "Testing")
plt.legend()

plt.show()
```

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(20)

plt.plot(epochs[0:50], trainAcc[0:50], label = "Training")
plt.plot(epochs[0:50], testAc[0:50], label = "Testing")
plt.legend()

plt.show()
```
